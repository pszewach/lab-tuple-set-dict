#### How many times does each letter in `letters` appear in `tup3`?

Print out the number of occurrence of each letter.

````python
# Your code here
number_of_ocurrences=[]
for letter in letters:
    number_of_ocurrences.append(tup3.count(letter))
print(number_of_ocurrences)

dict_ocurrences={letters[i]:number_of_ocurrences[i] for i in range (len(letters))}
print(dict_ocurrences)

````
